```python
# #Kütüphaneleri projeye ekleme
import tkinter
from tkinter import *
from PIL import ImageTk, Image
import cv2
import numpy as np
from keras.models import load_model
from PIL import Image

emotion_dict = {0: "Angry", 1: "Disgust", 2: "Fear", 3: "Happy", 4: "Sad", 5: "Surprise", 6: "Neutral"}


master = Tk()

canvas=Canvas(master,height=900,width=1200,bg='#0B1735')
canvas.pack()
Label(master,text='ethem').pack()
frame_ust=Frame(master,bg='#0B1735')
frame_ust.place(relx=0.1,rely=0.1,relwidth=0.8,relheight=0.08)
frame_sag=Frame(master,bg='#D9C9C9')
frame_sag.place(relx=0.6,rely=0.3,relwidth=0.3,relheight=0.6)
L1=Label()
L1.pack()
L2=Label(frame_ust,text="Gerçek Zamanlı Emoji Oluşturucu",bg='#0B1735',font=("Arial",40),fg='#ffffff')
L2.pack()
# frame_sol=Frame(master,bg='#D9C9C9')
# frame_sol.place(relx=0.1,rely=0.3,relwidth=0.3,relheight=0.6)

model = load_model("model.h5")

cap = cv2.VideoCapture(0)

while True:
    
    ret, frame = cap.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 140, 255), 2)
        roi_gray = gray[y:y + h, x:x + w]
        cropped_img = np.expand_dims(np.expand_dims(cv2.resize(roi_gray, (48, 48)), -1), 0)
        cv2.normalize(cropped_img, cropped_img, alpha=0, beta=1, norm_type=cv2.NORM_L2, dtype=cv2.CV_32F)
        prediction = model.predict(cropped_img)
        cv2.putText(frame, emotion_dict[int(np.argmax(prediction))], (x, y), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 140, 255), 2, cv2.LINE_AA)
        
        #print(emotion_dict[int(np.argmax(prediction))])
        
        if(emotion_dict[int(np.argmax(prediction))]=="Happy"):
            image1 = Image.open("happy.png")
            test = ImageTk.PhotoImage(image1)
            label1 = tkinter.Label(image=test)
            label1.image = test
            label1.place(relx=0.6,rely=0.3,relwidth=0.3,relheight=0.66)
            
        elif(emotion_dict[int(np.argmax(prediction))]=="Angry"):
            
            image1 = Image.open("angry.png")
            test = ImageTk.PhotoImage(image1)
            label1 = tkinter.Label(image=test)
            label1.image = test
            label1.place(relx=0.6,rely=0.3,relwidth=0.3,relheight=0.6)
        
        elif(emotion_dict[int(np.argmax(prediction))]=="Fear"):
            
            image1 = Image.open("fear.png")
            test = ImageTk.PhotoImage(image1)
            label1 = tkinter.Label(image=test)
            label1.image = test
            label1.place(relx=0.6,rely=0.3,relwidth=0.3,relheight=0.6)
        
        elif(emotion_dict[int(np.argmax(prediction))]=="Surprise"):
            
            image1 = Image.open("suprise.png")
            test = ImageTk.PhotoImage(image1)
            label1 = tkinter.Label(image=test)
            label1.image = test
            label1.place(relx=0.6,rely=0.3,relwidth=0.3,relheight=0.6)
        
        elif(emotion_dict[int(np.argmax(prediction))]=="Sad"):
            
            image1 = Image.open("sad.png")
            test = ImageTk.PhotoImage(image1)
            label1 = tkinter.Label(image=test)
            label1.image = test
            label1.place(relx=0.6,rely=0.3,relwidth=0.3,relheight=0.6)
        
        elif(emotion_dict[int(np.argmax(prediction))]=="Disgust"):
            
            image1 = Image.open("disgust.png")
            test = ImageTk.PhotoImage(image1)
            label1 = tkinter.Label(image=test)
            label1.image = test
            label1.place(relx=0.6,rely=0.3,relwidth=0.3,relheight=0.6)
            
            
         
            

    cv2.imshow('L1', frame)
    #L1['image']=frame
    master.update()
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from keras.models import load_model
emotion_dict = {0: "Angry", 1: "Disgust", 2: "Fear", 3: "Happy", 4: "Sad", 5: "Surprise", 6: "Neutral"}

model = load_model("video.h5")

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 1)
        roi_gray = gray[y:y + h, x:x + w]
        cropped_img = np.expand_dims(np.expand_dims(cv2.resize(roi_gray, (48, 48)), -1), 0)
        cv2.normalize(cropped_img, cropped_img, alpha=0, beta=1, norm_type=cv2.NORM_L2, dtype=cv2.CV_32F)
        prediction = model.predict(cropped_img)
        cv2.putText(frame, emotion_dict[int(np.argmax(prediction))], (x, y), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 1, cv2.LINE_AA)

    cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
